```python
# Divide a dataset into training, validation, test set by random sampling (randomness included).
def divide_dataset(N, test_prop = 0.1, val_prop = 0.1):
    
    # Get sample numbers for test, validation, and training.
    test_N = round(N * test_prop)
    val_N = round(N * val_prop)
    train_N = N - test_N - val_N

    # Sampling test index.
    test_index = np.sort(np.random.choice(N, test_N, replace=False))

    # Get trainval index by removing test index from all index.
    trainval_index = np.setdiff1d(np.array(range(N)), test_index)

    # Sampling validation index.
    val_index = np.sort(np.random.choice(trainval_index, val_N, replace = False))

    # Get train index by removing val index from trainval index.
    train_index = np.setdiff1d(trainval_index, val_index)

    # Summarize as dictionary.
    result = {"train_ids":train_index,"val_ids":val_index,"test_ids":test_index}

    return result

# Divide a dataset into training and test set by random sampling (randomness included).
def train_test_split(N, test_prop = 0.2):
    
    # Get sample sizes for training and test.
    test_N = round(N * test_prop)
    train_N = N - test_N
    
    # Sampling test index.
    test_index = np.sort(np.random.choice(N, test_N, replace=False))
    
    # Get train index by removing test index.
    train_index = np.setdiff1d(np.array(range(N)), test_index)
    
    result = {"train_ids":train_index, "test_ids":test_index}
    return result

# Get a sub-dataset of the given X, and Y by bagging.
def bagging(X, Y):
    
    N = X.shape[0]
    ids = np.random.choice(N, N, replace=True) # bagging.
    
    return X[ids], Y[ids]

# Get a sub-dataset of the given X, and Y by pasting.
def pasting(X, Y, prop = 0.6):
    
    N = X.shape[0]
    sample_N = round(N * prop)
    
    ids = np.random.choice(N, sample_N, replace=False) # pasting.
    
    return X[ids], Y[ids]
```
